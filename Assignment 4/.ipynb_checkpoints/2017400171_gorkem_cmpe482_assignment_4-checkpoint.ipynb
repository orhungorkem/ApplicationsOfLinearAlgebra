### a.  

$$ A = 
\begin{bmatrix}
1 & 6 \\
0 & -1 \\
-1 & -2
\end{bmatrix}
$$  

We are looking for SVD of $A$. 

$$ A = U.\Sigma.V^T $$  
where $U$ and $V$ are orthonormal matrices.  

Knowing that $A^T.A$ is a symmetric matrix, it can bi diagonalized with eigenvalue decomposition.  

$$ A^T.A = V.\Sigma^T.U^T.U.\Sigma.V^T = P.D.P^T$$  

$$ A^T.A = V.\Sigma^T.I.\Sigma.V^T = P.D.P^T $$  

$$ A^T.A = V.\Sigma^T.\Sigma.V^T = P.D.P^T $$

since $U$ is orthonormal and where $D$ and $\Sigma$ are diagonal matrices.  

See that $P = V$ here.

Let $\lambda_i$ be the ith diagonel index in $D$. And let $\sigma_i$ be the ith diagonal index in $\Sigma$. Having  

$$ \Sigma^T.\Sigma = D $$  
$$ \lambda_i = {\sigma_i}^2 $$   

for all $i$. Also note that size of $\Sigma$ is the same with size of $A$.

```python
A = np.array([[1,6],[0,-1],[-1,-2]])
D,P = np.linalg.eig(A.T@A)   #D=eigenvalues  P=eigenvectors
D
```

See the order of singular values is not descending. So we need to arrange our outcome in descending and consistent order (so also change P) to have correct form. The order is important in case of approximations.

```python
temp = D[0]
D[0] = D[1]
D[1] = temp

V = np.zeros(P.shape)
for i in range(len(P)):
    V[i,0] = P[i,1]
    V[i,1] = P[i,0]

D = [math.sqrt(i) for i in D]
Sigma = np.zeros(A.shape)
for i in range(len(D)):
    Sigma[i,i]=D[i]
```

Now consider in case of $U$ and revert the procedure.  

$$ A.A^T = U.\Sigma.V^T. V.\Sigma^T.U^T = P.D.P^T$$  

$$ A.A^T = U.\Sigma.I.\Sigma^T.U^T = P.D.P^T $$  

$$ A.A^T = U.\Sigma.\Sigma^T.U^T = P.D.P^T $$  

since $V$ is orthonormal and where $D$ and $\Sigma$ are diagonal matrices.  

See that $P = U$ here.

```python
D,P = np.linalg.eig(A@A.T)   #D=eigenvalues  P=eigenvectors

U = P
```

Checking

```python
U@Sigma@V.T
```

See it brings a matrix very close to $A$. Now compare the components with the built-in numpy function results.

```python
u,s,v = np.linalg.svd(A)
sigma = np.zeros(A.shape)
for i in range(len(s)):
    sigma[i,i]=s[i]
```

### b.  

For $A$ with real elements, we can be sure that there must exist an orthogonal $U$ and $V$. This is because $U$ and $V$ correspond to $P$ matrices in eigendecompositions of $A^T.A$ and $A.A^T$ as explained in the question above. But have can we make sure of resulting $P$'s from eigendecompositions of $A^T.A$ and $A.A^T$ being orthonormal?   

The answer comes from Spectral Theorem. We know that matrix $P$ is the collection of eigenvectors of given matrix. Spectral theorem asserts that if $A \in R_{n×n}$ is symmetric, there exists an orthonormal basis of the corresponding vector space $V$ consisting of eigenvectors of $A$, and each eigenvalue is real. (MML)   

In our case, we can say that if $A^T.A$ and $A.A^T$ have eigenvectors which construct orthonormal bases, they can be decomposed into $P.D.P^T$, thus there must exist orthogonal(orthonormality implies orthogonality) $U$ and $V$ resulting from SVD of $A$. With Spectral Theorem, we can claim that if $A^T.A$ and $A.A^T$ are symmetric, there must exist orthogonal $U$ and $V$ resulting from SVD of $A$. (Using transitivity)  

We already know that $A^T.A$ and $A.A^T$ are symmetric for any real valued $A$ since

$$(A^T.A)^T = A^T.A$$
$$(A.A^T)^T = A.A^T$$  

Hence, for $A$ with real elements, we can be sure that there must exist an orthogonal $U$ and $V$.


### c.

Let $\hat{A}$ be the rank one approximation of $A$. Then 

$$\hat{A} = u_1.\sigma_1.{v_1}^T $$  

where $u_1$ and $v_1$ are the first columns of $U$ and $V$.

```python
mul1 = np.array([[U[i,0]] for i in range(len(U))])
mul2 = np.array([[V[0,i] for i in range(len(V[0]))]])
A_hat = mul1@mul2
A_hat = Sigma[0,0]*A_hat
A_hat
```

According to Eckhart-Young Theorem, the norm of the difference between the original matrix and its rank k approximation is equal to the kth largest singular value.  

$${||A-\hat{A}||}_2 = \sigma_2 $$

```python
print("Norm: "+str(np.linalg.norm(A_hat-A)))
print("Second singular value: "+str(math.sqrt(D[1])))
print("Difference: "+str(np.linalg.norm(A_hat-A)-math.sqrt(D[1])))
```

This result verifies the theorem.

$$\Sigma = \begin{bmatrix}

```python
def create_patches(A):
    B = []
    for i in range(9):
        array = []
        for j in range(9):
            array = A[i*7:i*7+7,j*7:j*7+7]
            array = array.flatten()
            B.append(array)
            
    
    return np.array(B)

def create_image_from_patches(B):
    arrays = []
    for i in B:
        arrays.append(np.reshape(i,(7,7)))
        
    A =np.zeros((63,63))
    for i in range(9):
        for j in range(9):
            array = arrays[i*9+j]  # a 7x7 array
            for k in range(7):
                for m in range(7):
                    A[i*7+k][j*7+m] = array[k][m]
    return A
        
```

```python
B1 = create_patches(A_1)
B2 = create_patches(A_2)
B3 = create_patches(A_3)

plt.imshow(A_1, cmap="gray");
plt.figure()
plt.imshow(create_patches(A_1), cmap="gray");
plt.figure()
plt.imshow(create_image_from_patches(create_patches(A_1)), cmap="gray");
plt.figure()
```

Having the first column with low variance gives first columns a summary property. Consider the following, according to SVD theorem,  

$$ A.V = U.\Sigma $$ ,
in our case, $A$ is actually $B_1$. Then,  

$$ B_1.{v_1}_1 = {u_1}_1.\Sigma_1 $$  
where ${v_1}_1$ and ${u_1}_1$ are first columns of $U_1 $ and $V_1$.  

Think of matrix multiplication, if ${v_1}_1$ has low variance, then all indices in all rows of $B_1$ will be multiplied with similar weights. Since $B_1$ consists of rows that are corresponding to 7x7 grids in our original image, multiplication with  ${v_1}_1$ means getting an average of each 7x7 grid in some sense. So we capture an average behavior of the image with that first column, and so with first rank approximation.  

Consider the other cases, for example in second column of $V_2$, weights change periodically. Apparently, this column will capture some other specific characteristics of given image, of course it will be multiplied with a smaller $\sigma$, this helps to keep details as **details**.

This helps to see the logic of approximation. With first $V$ column, we generate a very blurred image with only the average behavior. The we add details step by step and improve our approximations. 

### e.

```python
def approximate(U,V,S,rank):
    
    mul1 = U[:,:rank]
    mul2 = S[:rank,:rank]
    mul3 = V[:rank,:]
    return mul1@mul2@mul3


S1 = np.zeros((len(B1),len(B1[0])))
for i in range(len(Sigma1)):
    S1[i,i]= Sigma1[i]

f, ax = plt.subplots(1,4,figsize = (20,5))
for j,k in zip(range(4),[1,5,10,20]):
    m = approximate(U1,V1,S1,k)
    ax[j].set_title("Rank: "+str(k))
    ax[j].imshow(create_image_from_patches(m), cmap="gray");
    
    
    
    
```

### f. 

Such approximations save us a huge space since we can regenerate our images by just storing a couple of rows, columns and singular values. In a rank $r$ approximation, we simply keep $r$ columns from $U$, $r$ singular values, and $r$ rows from $V^T$. 
If we decompose a matrix $A$ with size $m x n$, we will store $m.r + r + r.n$ pixels in a rank r approximation.
In our case:  

* Rank 1: $81.1+1+1.49 = 131$ pixels stored. We keep $\frac{131}{3969} \approx 0.033$ of the parameters.
* Rank 5: $81.5+5+5.49 = 655$ pixels stored. We keep $\frac{655}{3969} \approx 0.165$ of the parameters.
* Rank 10: $81.10+10+10.49 = 1310$ pixels stored. We keep $\frac{1310}{3969} \approx 0.33$ of the parameters.
* Rank 20: $81.20+20+20.49 = 2620$ pixels stored. We keep $\frac{2620}{3969} \approx 0.66$ of the parameters.
