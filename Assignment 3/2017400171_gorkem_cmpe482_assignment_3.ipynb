### a.
Find and print $\hat{\theta_1}$ and $\hat{\theta_2}$ using regularized least squares, by setting $\lambda$ equal to .00001 and 100000, respectively. Compare $\hat{\theta_1}$ and $\hat{\theta_2}$ to the original $\hat{\theta}$ you found in Q1. Which
one is closer to $\hat{\theta}$? What does the other one look like? Why?


Now the problem becomes minimizing,
$$ ||X.\theta-y||^{2} + \lambda.||\theta||^{2} $$
$$ = ||X.\theta-y||^{2} + \lambda.||I.\theta-0||^{2} $$  

Then form $\bar{X}$ and $\bar{y}$ such that,  
$$ \bar{X} =\begin{bmatrix}
X \\
\sqrt{\lambda}.I
\end{bmatrix}, \bar{y} =\begin{bmatrix}
y \\
0
\end{bmatrix}$$  
Now minimizing
$$ J = || \bar{X}.\theta -\bar{y} ||^2 $$
is the single criterion problem we can easily solve.  
$$ \hat{\theta} = ({\bar{X}}^{T}.\bar{X})^{-1}.(\bar{X})^{T}.\bar{y} $$

```python
lambda1 = 0.00001
j2 = math.sqrt(lambda1)*np.identity(8)
X_bar = np.concatenate((X_train,j2),axis = 0)
y2 = np.zeros(8)
y_bar = np.concatenate((y_train,y2),axis = 0)

theta1 = np.linalg.inv(X_bar.T @ X_bar) @ X_bar.T @ y_bar
theta1 
```

```python
lambda2 = 100000
j2 = math.sqrt(lambda2)*np.identity(8)
X_bar = np.concatenate((X_train,j2),axis = 0)
y2 = np.zeros(8)
y_bar = np.concatenate((y_train,y2),axis = 0)

theta2 = np.linalg.inv(X_bar.T @ X_bar) @ X_bar.T @ y_bar
theta2 
```

$\hat{\theta_1}$ is very close to $\hat{\theta}$. This is perfectly normal since 0.00001 is a very small regularization coefficient to have an effect on the solution to our least squares problem. In other words, 0.00001 is not a strong incentive to decrease the coefficients. $\hat{\theta_2}$ has very low values since 100000 is a very large regularization coefficient and increasing $\theta$ brings an effective punishment.

### b. 
Find the value of that is best for generalization $\lambda$ among the candidates {0,.001,.01,.1,1.}. Print the training RMS and test RMS for all values of $\lambda$, and clearly state your chosen $\lambda$ as the best one.


```python
def solve_regularized_problem(X_train, y_train,lamb):
    
    j2 = math.sqrt(lamb)*np.identity(len(X_train[0]))
    X_bar = np.concatenate((X_train,j2),axis = 0)
    y2 = np.zeros(len(X_train[0]))
    y_bar = np.concatenate((y_train,y2),axis = 0)

    return np.linalg.inv(X_bar.T @ X_bar) @ X_bar.T @ y_bar
    
for i in [0,0.001,0.01,0.1,1]:
    theta = solve_regularized_problem(X_train,y_train,i)
    print(theta)
    pred = X @ theta
    error = pred - y
    rms_train = np.linalg.norm(error[0:600])/math.sqrt(600)
    rms_test = np.linalg.norm(error[600:])/math.sqrt(174)
    print("***************************")
    print("Lambda = "+ str(i))
    print("Train rms error: "+ str(rms_train))
    print("Test rms error: "+ str(rms_test))
    
```

Although there is no major difference among results, test rms error increases more that train rms error as we increase $\lambda$. Therefore, 0 regularization gives the best generalization performance. This means regularization has no benefit when there is no overfitting. In our case we can say that there is no overfitting because train and test set errors are close to each other.

* For $\lambda>0$, $M$ is always invertible since 
$$  M = {\hat{X}}^T.\hat{X} + \lambda.I = \bar{X}^T.\bar{X} $$
where
$$\bar{X} =
\begin{bmatrix}
X \\
\sqrt{\lambda}.I
\end{bmatrix}
$$

    and $\bar{X}$ always have linearly independent columns for $\lambda>0$. (For any sized $X$) Therefore, there is always a unique solution.

* For $\lambda = 0$,
$$M={\hat{X}}^T.\hat{X} + \lambda.I = {\hat{X}}^T.\hat{X}$$
We know that $\hat{X}$ is not invertible since it is a 30x36 matrix and must have linearly dependent columns by Pigeonhole Principle. Thus, $ M =  {\hat{X}}^T.\hat{X}$ is also not invertible. Then we can not find a solution using this formula for $\lambda=0$.

* As alternative, we can use right inverse of $\hat{X}$. See that when $\lambda=0$, the equation reduces to,

    $$ \theta = ({\hat{X}}^T.\hat{X})^{-1}. {\hat{X}}^T.\hat{y} $$

    this corresponds to left inverse of $\hat{X}$. SÄ°nce $\hat{X}$ is a wide matrix, we can not apply this but we can apply the right inverse. Consider the following equation:  
    
    $$ \theta =  {\hat{X}}^T.(\hat{X}.{\hat{X}}^T)^{-1}.\hat{y} $$
    $$ \hat{X}.\theta =  (\hat{X}.{\hat{X}}^T).(\hat{X}.{\hat{X}}^T)^{-1}.\hat{y} $$
    $$ \hat{X}.\theta = \hat{y} $$  
    
    As you see, we were able to reach a solution. This is actually a clear solution since we are able to find $\theta$ to fit perfectly to the given train set when the system is under determined as in this case.

$C=$

$$ \begin{bmatrix}

0&0&0&0&0 &1 & 0.5 & {0.5}^2 & {0.5}^3 & {0.5}^4 & -1 & -(0.5) & -{(0.5)}^2 & -{(0.5)}^3 & -{(0.5)}^4   \\
 0&  1 & 2.{(-0.5)} & 3.{(-0.5)}^2 & 4.{(-0.5)}^3 & 0 & -1 & -2.{(-0.5)} & -3.{(-0.5)}^2 & -4.{(-0.5)}^3 &0&0&0&0&0 \\

$$ d = \begin{bmatrix} 0\\0\\0\\0 \end{bmatrix} $$

### c.

```python
A = np.zeros((150,15))
for i in range(150):
    if x[i]<-0.5:
        for j in range(5):
            A[i][j]=x[i]**j
            
    elif x[i] <0.5:
        for j in range(5):
            A[i][5+j]=x[i]**j
    
    else:
        for j in range(5):
            A[i][10+j]=x[i]**j
    
C =np.zeros((6,15))

for i in range(5):
    C[0][i]=(-0.5)**i
    C[0][i+5]=-1*(-0.5)**i
    C[1][i+5]=(0.5)**i
    C[1][i+10]=-1*(0.5)**i
    
for i in range(4):
    C[2][i+1]=(i+1)*(-0.5)**i
    C[2][i+6]=-1*(i+1)*(-0.5)**i
    C[3][i+6]=(i+1)*(0.5)**i
    C[3][i+11]=-1*(i+1)*(0.5)**i
    
for i in range(3):
    C[4][i+2]=(i+1)*(i+2)*(-0.5)**i
    C[4][i+7]= -(i+1)*(i+2)*(-0.5)**i
    C[5][i+7]=(i+1)*(i+2)*(0.5)**i
    C[5][i+12]= -(i+1)*(i+2)*(0.5)**i
    
    
d = np.zeros(6)
    
```

### d. 

Now we are going to solve the problem with the following algorithm: 

$$ \begin{bmatrix} A \\ C \end{bmatrix} = \begin{bmatrix} Q_1 \\ Q_2 \end{bmatrix}.R $$

$$ {Q_2}^{T} = \tilde{Q}.\tilde{R}$$

$$ w = {\tilde{R}}^{-1}.(2.{\tilde{Q}}^{T}.{Q_1}^{T}.y - 2.{\tilde{R}}^T.d)$$

$$ c = R^{-1}.({Q_1}^{T}.y - \frac{1}{2}.{Q_2}^{T}.w) $$

```python
Q, R = np.linalg.qr(np.vstack([A,C]))
Q1 = Q[0:150,:]
Q2 = Q[150:,:]
Q_tilde, R_tilde = np.linalg.qr(Q2.T)
w = np.linalg.inv(R_tilde) @ (2* Q_tilde.T @ Q1.T @ y)
c = np.linalg.inv(R) @ (Q1.T @ y - 0.5* Q2.T @ w)

plt.stem(c)
```

### e.

```python
y_approx = A @ c
f = plt.figure(figsize=(8, 8))
plt.plot(x,y_approx)
plt.scatter(x,y,color='r')
```

### f.  

Now we need to verify that our solution satisfies KKT conditions.

$$ \begin{bmatrix} 2.A^TA & C^T \\ C & 0 \end{bmatrix}.\begin{bmatrix} c \\ z \end{bmatrix} = \begin{bmatrix} 2.A^T.b \\ d \end{bmatrix}$$

As we solve this equation for c and z, we get:

```python
temp1 = np.vstack([2*A.T@A,C])
temp2 = np.vstack([C.T,np.zeros((6,6))])
M = np.column_stack([temp1,temp2])

K = np.concatenate((2*A.T@y,d))

solution = np.linalg.inv(M)@K
solution
```

```python
c_validate = plt.stem(solution[0:15])
c_validate
```

```python
z = solution[15:]
plt.stem(z)
```

See that we arrived at the same c and we got a valid z from solving the KKT equation.
